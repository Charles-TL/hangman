```python
import random
import string

# I have downloaded a words list from the website "hangmanwords.com" as a txt file
# Now opening the txt file and stripping and spliting to convert it into a list
list_file = open("Documents/Programming/WordList.txt","r")
new_list = []
for line in list_file:
    stripped_line = line.strip()
    line_list = stripped_line.split()
    new_list.append(line_list)
list_file.close()

#print(new_list)
```

```python
# Converting new_list into a single list
wordlist = []
for word in new_list:
    wordlist += word
#print(wordlist)
```

```python
# Let's pick a random word in wordlist
def random_word(wordlist):
    guess_word = random.choice(wordlist)
    return guess_word
```

```python
# Defining the rules
def hangman():
    word = random_word(wordlist) # Word to guess
    letters_in_word = set(word)  # letters in the word
    alphabet = set(string.ascii_lowercase) # using ascii alphabet from the string library
    used_letters = set()  # This shall be the letters already used

    lives = 10

    # Player input
    while len(letters_in_word) > 0 and lives > 0:

        print('You now have', lives, 'lives left and you have used these letters so far: ', ' '.join(used_letters))

        # Displaying guessword with "-" for letters not yet found
        guessword = [letter if letter in used_letters else '-' for letter in word]
        print('Guess word: ', ' '.join(guessword))

        player_letter = input('Guess a letter: ').lower()
        if player_letter in used_letters:
            print('\nYou have already used that letter. Pick another one.')
        elif player_letter in alphabet - used_letters:
            used_letters.add(player_letter)
            if player_letter in letters_in_word:
                letters_in_word.remove(player_letter)
                print('') #leaves a line before the next attempt

            else:
                lives = lives - 1  # takes one life away if wrong choice
                print('\nYour letter,', player_letter, 'is not in the word.')
        else:
            print('\nNot valid. You can pick only one letter from the alphabet.')

    # conclusion of the game, when len(letters_in_word) == 0 OR lives == 0
    if lives == 0:
        print('You lost. The word was', word)
    else:
        print('Well done!', word, ' was the word indeed.')


if __name__ == '__main__':
    hangman()
```
